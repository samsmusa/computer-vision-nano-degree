```python

import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

# Define gaussian, sobel, and laplacian (edge) filters
```

```python
gaussian = (1/9)*np.array([
    [1,1,1],
    [1,1,1],
    [1,1,1]
])
sobel_x = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1],
])
sobel_y = sobel_x.T
laplacian = np.array([
    [0, 1, 0],
    [1, -4, 1],
    [0, 1, 0],
])
filter_name = ['gaussian', 'sobel_x', 'sobel_y', 'laplacian']
filters = [eval(filter_n) for filter_n in filter_name]
```

```python
f_filters = [np.fft.fft2(x) for x in filters]
f_shift = [np.fft.fftshift(y) for y in f_filters]
frequency_tx = [np.log(np.abs(z)+1) for z in f_shift]

for i in range(len(filters)):
    plt.subplot(2, 2, i+1)
    plt.imshow(filters[i], cmap="gray")
    plt.title(filter_name[i])
plt.show()
```
